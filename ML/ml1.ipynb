```python
import numpy as np

def distance_transform(longitude1, latitude1, longitude2, latitude2):
    # Convert degrees to radians
    long1, lat1, long2, lat2 = map(np.radians, [longitude1, latitude1, longitude2, latitude2])
    
    # Calculate differences between longitudes and latitudes
    dist_long = long2 - long1
    dist_lat = lat2 - lat1
    
    # Apply Haversine formula
    a = np.sin(dist_lat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dist_long / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radius of Earth in kilometers (mean radius)
    r = 6371.0
    
    # Calculate the distance
    distance = c * r
    
    return distance
```

```python
# Assuming distance_transform is a function you have defined
df['Distance'] = distance_transform(
    df['pickup_longitude'],
    df['pickup_latitude'],
    df['dropoff_longitude'],
    df['dropoff_latitude']
)
```
